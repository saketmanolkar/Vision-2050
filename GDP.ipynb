```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Assuming your DataFrame is named df

# Select data for India
df_india = df[df['Country Name'] == 'India']
df_india = df_india.dropna()

# Extract years and GDP values without specifying column names
years_india = df_india.columns[1:].values.astype(int)
gdp_india = df_india.iloc[:, 1:].apply(pd.to_numeric, errors='coerce').values.flatten()

# Select data for first-world countries
first_world_countries = ['United States', 'Canada', 'United Kingdom', 'Germany', 'France', 'Japan', 'Australia']
df_first_world = df[df['Country Name'].isin(first_world_countries)]
df_first_world = df_first_world.dropna()

# Extract years and GDP values without specifying column names
years_first_world = df_first_world.columns[1:].values.astype(int)
gdp_first_world = df_first_world.iloc[:, 1:].apply(pd.to_numeric, errors='coerce').values

# Handle missing values by filling with the mean
gdp_india = np.nan_to_num(gdp_india, nan=np.nanmean(gdp_india))
gdp_first_world = np.nan_to_num(gdp_first_world, nan=np.nanmean(gdp_first_world))

# Filter data for training on years 1990 to 2022
start_year_training = 1990
end_year_training = 2022
mask_training = (years_india >= start_year_training) & (years_india <= end_year_training)

# Apply the mask for both India and first-world countries
years_india_training = years_india[mask_training]
gdp_india_training = gdp_india[mask_training]

# Calculate the average GDP for each year across first-world countries
avg_gdp_first_world = np.mean(gdp_first_world, axis=0)

# Set a larger plot size
plt.figure(figsize=(12, 6))

# Plot actual values for India
plt.plot(years_india, gdp_india, label='Actual GDP (India)', marker='o', linestyle='-')

# Plot actual values for the average of first-world countries
plt.plot(years_first_world, avg_gdp_first_world, label='Actual Avg. GDP (First World)', color='r', linestyle='--')

# ARIMA order selection (You can experiment with different values)
order_india = (1, 1, 3)
order_first_world = (2, 1, 1)

# ARIMA for India with different order
model_india = ARIMA(gdp_india_training, order=order_india)
fit_india = model_india.fit()
future_years_india = np.arange(2023, 2051)
predicted_gdp_india = fit_india.forecast(steps=len(future_years_india))

# ARIMA for the average of first-world countries with different order
model_first_world = ARIMA(avg_gdp_first_world, order=order_first_world)
fit_first_world = model_first_world.fit()
predicted_gdp_first_world = fit_first_world.forecast(steps=len(future_years_india))

# Print predicted values for India
print("Predicted GDP (India):", predicted_gdp_india)

# Print predicted values for the average of first-world countries
print("Predicted GDP (First World):", predicted_gdp_first_world)

# Calculate and print accuracy measures
mae_india = mean_absolute_error(gdp_india[-len(future_years_india):], predicted_gdp_india)
mse_india = mean_squared_error(gdp_india[-len(future_years_india):], predicted_gdp_india)

mae_first_world = mean_absolute_error(avg_gdp_first_world[-len(future_years_india):], predicted_gdp_first_world)
mse_first_world = mean_squared_error(avg_gdp_first_world[-len(future_years_india):], predicted_gdp_first_world)

print("MAE (India):", mae_india)
print("MSE (India):", mse_india)

print("MAE (First World):", mae_first_world)
print("MSE (First World):", mse_first_world)

# Plot predicted values for India
plt.plot(future_years_india, predicted_gdp_india, label='ARIMA Prediction (India)', linestyle='--', color='blue')

# Plot predicted values for the average of first-world countries
plt.plot(future_years_india, predicted_gdp_first_world, label='ARIMA Prediction (First World)', linestyle='--', color='green')

# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('GDP Comparison: India vs. First World Countries (ARIMA)')
plt.legend()

# Increment x-axis ticks by 5
plt.xticks(range(1990, 2051, 5), rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

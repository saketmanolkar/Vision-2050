```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from statsmodels.tsa.arima.model import ARIMA

# Assuming 'df' is your DataFrame with relevant data
# Replace 'NA' with NaN
df.replace('NA', np.nan, inplace=True)

# Filter data for India
india_data = df[df['Country Name'] == 'India'].iloc[:, 1:].astype(float)

# Fill missing values using the mean
india_data_filled = india_data.fillna(india_data.mean())

# Set a larger figure size
plt.figure(figsize=(12, 6))

# Plotting the literacy rate for India (Actual)
years_actual_india = india_data_filled.columns.astype(int)
values_actual_india = india_data_filled.values.flatten()
plt.plot(years_actual_india, values_actual_india, marker='o', label='India (Actual)')

# Adding labels and title
plt.title('Literacy Rate in India Over the Years')
plt.xlabel('Year')
plt.ylabel('Literacy Rate')
plt.grid(True)

# Linear Regression
years_linear_regression = np.array(years_actual_india).reshape(-1, 1)
values_actual_india_no_nan = values_actual_india[~np.isnan(values_actual_india)]  # Remove NaN values for fitting
years_linear_regression_no_nan = years_linear_regression[~np.isnan(values_actual_india)]
model_linear = LinearRegression()
model_linear.fit(years_linear_regression_no_nan, values_actual_india_no_nan)
linear_regression_predictions = model_linear.predict(years_linear_regression)

# Plotting Linear Regression predictions from 2023 to 2050
years_future = np.arange(2023, 2051).reshape(-1, 1)
linear_regression_predictions_future = model_linear.predict(years_future)
plt.plot(np.arange(2023, 2051), linear_regression_predictions_future, label='Linear Regression (Predicted)', linestyle='--')

# Calculate and print R-squared for Linear Regression
r2_linear = model_linear.score(years_linear_regression_no_nan, values_actual_india_no_nan)
print(f"R-squared for Linear Regression: {r2_linear}")

# ARIMA
order_arima = (1, 1, 1)  # You may need to adjust the order based on the characteristics of your data
model_arima = ARIMA(values_actual_india, order=order_arima)
fitted_model_arima = model_arima.fit()  # Use the default 'lbfgs' method
arima_predictions = fitted_model_arima.predict(start=len(values_actual_india), end=len(values_actual_india) + 27)  # Predict from 2023 to 2050
plt.plot(np.arange(2023, 2051), arima_predictions, label='ARIMA (Predicted)', linestyle='--')

# Calculate and print R-squared for ARIMA
residuals_arima = values_actual_india - fitted_model_arima.fittedvalues
ssr_arima = np.sum(residuals_arima**2)
sst_arima = np.sum((values_actual_india - np.mean(values_actual_india))**2)
r2_arima = 1 - (ssr_arima / sst_arima)
print(f"R-squared for ARIMA: {r2_arima}")

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

# Assuming 'df' is your DataFrame with relevant data
# Replace 'NA' with NaN
df.replace('NA', np.nan, inplace=True)

# Filter data for India
india_data = df[df['Country Name'] == 'India'].iloc[:, 1:].astype(float)

# Fill missing values using the mean
india_data_filled = india_data.fillna(india_data.mean())

# Set a larger figure size
plt.figure(figsize=(12, 6))

# Plotting the literacy rate for India (Actual)
years_actual_india = india_data_filled.columns.astype(int)
values_actual_india = india_data_filled.values.flatten()
plt.plot(years_actual_india, values_actual_india, marker='o', label='India (Actual)')

# Adding labels and title
plt.title('Literacy Rate in India Over the Years')
plt.xlabel('Year')
plt.ylabel('Literacy Rate')
plt.grid(True)

# Linear Regression
years_linear_regression = np.array(years_actual_india).reshape(-1, 1)
values_actual_india_no_nan = values_actual_india[~np.isnan(values_actual_india)]  # Remove NaN values for fitting
years_linear_regression_no_nan = years_linear_regression[~np.isnan(values_actual_india)]
model_linear = LinearRegression()
model_linear.fit(years_linear_regression_no_nan, values_actual_india_no_nan)
linear_regression_predictions = model_linear.predict(years_linear_regression)

# Polynomial Regression with Degree 2
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(years_linear_regression_no_nan)
model_poly = LinearRegression()
model_poly.fit(X_poly, values_actual_india_no_nan)
poly_predictions = model_poly.predict(poly_features.transform(years_linear_regression_no_nan))

# Print R-squared for Polynomial Regression (Degree 2)
r2_poly = r2_score(values_actual_india_no_nan, poly_predictions)
print(f"R-squared for Polynomial Regression (Degree 2): {r2_poly}")

# Print the predicted values
poly_predictions_future = model_poly.predict(poly_features.transform(np.arange(2023, 2051).reshape(-1, 1)))
print("Predicted Values for the Years 2023 to 2050 (Polynomial Regression Degree 2):")
for year, prediction in zip(np.arange(2023, 2051), poly_predictions_future):
    print(f"Year {year}: {prediction}")

# Plotting Polynomial Regression predictions from 2023 to 2050
plt.plot(np.arange(2023, 2051), poly_predictions_future, label='Polynomial Regression (Degree 2) Predicted', linestyle='--')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

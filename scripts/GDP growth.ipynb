```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Assuming your DataFrame is named df

# Select data for India
df_india = df[df['Country Name'] == 'India']

# Replace 'NA' with NaN and convert the DataFrame to numeric
df_india.replace('NA', np.nan, inplace=True)
df_india = df_india.apply(pd.to_numeric, errors='coerce')

# Extract years and GDP growth values
years_india = df_india.columns[1:].values.astype(int)
gdp_growth_india = df_india.iloc[:, 1:].values.flatten()

# Handle missing values by filling with the mean
gdp_growth_india = np.nan_to_num(gdp_growth_india, nan=np.nanmean(gdp_growth_india))

# Filter data for training on years 1961 to 2022
start_year_training = 1961
end_year_training = 2022
mask_training = (years_india >= start_year_training) & (years_india <= end_year_training)

# Apply the mask for India
years_india_training = years_india[mask_training]
gdp_growth_india_training = gdp_growth_india[mask_training]

# Set a larger plot size
plt.figure(figsize=(12, 6))

# Plot actual values for India
plt.plot(years_india, gdp_growth_india, label='Actual GDP Growth (India)', marker='o', linestyle='-')

# ARIMA order selection (You can experiment with different values)
order_india = (0, 1, 0)

# ARIMA for India with different order
model_india = ARIMA(gdp_growth_india_training, order=order_india)
fit_india = model_india.fit()
future_years_india = np.arange(2023, 2051)
predicted_gdp_growth_india = fit_india.forecast(steps=len(future_years_india))

# Print predicted values for India
print("Predicted GDP Growth (India):", predicted_gdp_growth_india)

# Calculate and print accuracy measures
mae_india = mean_absolute_error(gdp_growth_india[-len(future_years_india):], predicted_gdp_growth_india)
mse_india = mean_squared_error(gdp_growth_india[-len(future_years_india):], predicted_gdp_growth_india)

print("MAE (India):", mae_india)
print("MSE (India):", mse_india)

# Plot predicted values for India
plt.plot(future_years_india, predicted_gdp_growth_india, label='ARIMA Prediction (India)', linestyle='--', color='blue')

# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('GDP Growth')
plt.title('GDP Growth Prediction for India (ARIMA)')
plt.legend()

# Increment x-axis ticks by 5
plt.xticks(range(1960, 2051, 5), rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Assuming your DataFrame is named df

# Select data for India
df_india = df[df['Country Name'] == 'India'].copy()  # Make a copy to avoid SettingWithCopyWarning

# Replace 'NA' with NaN and convert the DataFrame to numeric
df_india.replace('NA', np.nan, inplace=True)
df_india = df_india.apply(pd.to_numeric, errors='coerce')

# Extract years and GDP growth values
years_india = df_india.columns[1:].values.astype(int)
gdp_growth_india = df_india.iloc[:, 1:].values.flatten()

# Handle missing values by filling with the mean
gdp_growth_india = np.nan_to_num(gdp_growth_india, nan=np.nanmean(gdp_growth_india))

# Filter data for training on years 1961 to 2022
start_year_training = 1961
end_year_training = 2022
mask_training = (years_india >= start_year_training) & (years_india <= end_year_training)

# Apply the mask for India
years_india_training = years_india[mask_training]
gdp_growth_india_training = gdp_growth_india[mask_training]

# Reshape the data for linear regression
X_india = years_india_training.reshape(-1, 1)

# Set a larger plot size
plt.figure(figsize=(12, 6))

# Plot actual values for India
plt.plot(years_india, gdp_growth_india, label='Actual GDP Growth (India)', marker='o', linestyle='-')

# Linear regression model for India
model_india = LinearRegression()
model_india.fit(X_india, gdp_growth_india_training)
predicted_gdp_growth_india_training = model_india.predict(X_india)

# Print predicted values for India
print("Predicted GDP Growth (India):", predicted_gdp_growth_india_training)

# Calculate and print accuracy measures
mae_india = mean_absolute_error(gdp_growth_india_training, predicted_gdp_growth_india_training)
mse_india = mean_squared_error(gdp_growth_india_training, predicted_gdp_growth_india_training)
r2_india = r2_score(gdp_growth_india_training, predicted_gdp_growth_india_training)

print("MAE (India):", mae_india)
print("MSE (India):", mse_india)
print("R-squared (India):", r2_india)

# Plot predicted values for India
plt.plot(years_india_training, predicted_gdp_growth_india_training, label='Linear Regression Prediction (India)', linestyle='--', color='blue')

# Predict values for the years 2023 to 2050
future_years_india = np.arange(2023, 2051).reshape(-1, 1)
predicted_gdp_growth_india_future = model_india.predict(future_years_india)

# Print predicted values for the future
print("Predicted GDP Growth (India) for 2023 to 2050:", predicted_gdp_growth_india_future)

# Plot predicted values for the future
plt.plot(future_years_india.flatten(), predicted_gdp_growth_india_future, label='Linear Regression Prediction (India, Future)', linestyle='--', color='orange')

# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('GDP Growth')
plt.title('GDP Growth Prediction for India (Linear Regression)')
plt.legend()

# Increment x-axis ticks by 5
plt.xticks(range(1960, 2051, 5), rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

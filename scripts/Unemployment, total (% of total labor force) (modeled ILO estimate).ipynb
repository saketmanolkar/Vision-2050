```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Assuming your DataFrame is named df

# Select data for India
df_india = df[df['Country Name'] == 'India']
df_india = df_india.dropna()

# Extract years and unemployment values without specifying column names
years_india = df_india.columns[1:].values.astype(int)
unemployment_india = df_india.iloc[:, 1:].apply(pd.to_numeric, errors='coerce').values.flatten()

# Select data for first-world countries
first_world_countries = ['United States', 'Canada', 'United Kingdom', 'Germany', 'France', 'Japan', 'Australia']
df_first_world = df[df['Country Name'].isin(first_world_countries)]
df_first_world = df_first_world.dropna()

# Extract years and unemployment values without specifying column names
years_first_world = df_first_world.columns[1:].values.astype(int)
unemployment_first_world = df_first_world.iloc[:, 1:].apply(pd.to_numeric, errors='coerce').values

# Filter data for training on years 1990 to 2022
start_year_training = 1990
end_year_training = 2022
mask_training = (years_india >= start_year_training) & (years_india <= end_year_training)

# Apply the mask for both India and first-world countries
years_india_training = years_india[mask_training]
unemployment_india_training = unemployment_india[mask_training]

# Calculate the average unemployment rate for each year across first-world countries
avg_unemployment_first_world = np.mean(unemployment_first_world, axis=0)

# Set a larger plot size
plt.figure(figsize=(12, 6))

# Plot actual values for India
plt.plot(years_india, unemployment_india, label='Actual Unemployment (India)', marker='o', linestyle='-')

# Plot actual values for the average of first-world countries
plt.plot(years_first_world, avg_unemployment_first_world, label='Actual Avg. Unemployment (First World)', color='r', linestyle='--')

# ARIMA order selection using auto_arima or other methods

# ARIMA for India
model_india = ARIMA(unemployment_india_training, order=(0, 1, 2))
fit_india = model_india.fit()
future_years_india = np.arange(2023, 2051)
predicted_unemployment_india = fit_india.forecast(steps=len(future_years_india))

# ARIMA for the average of first-world countries
model_first_world = ARIMA(avg_unemployment_first_world, order=(5, 1, 1))
fit_first_world = model_first_world.fit()
predicted_unemployment_first_world = fit_first_world.forecast(steps=len(future_years_india))

# Print predicted values only
print("Predicted Unemployment (India):", predicted_unemployment_india)
print("Predicted Unemployment (First World):", predicted_unemployment_first_world)

# Plot predicted values for India
plt.plot(future_years_india, predicted_unemployment_india, label='ARIMA Prediction (India)', linestyle='--', color='blue')

# Plot predicted values for the average of first-world countries
plt.plot(future_years_india, predicted_unemployment_first_world, label='ARIMA Prediction (First World)', linestyle='--', color='green')

# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('Unemployment %')
plt.title('Unemployment Rate Comparison: India vs. First World Countries (ARIMA)')
plt.legend()

# Increment x-axis ticks by 5
plt.xticks(range(1990, 2051, 5), rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Assuming your DataFrame is named df

# Select data for India
df_india = df[df['Country Name'] == 'India']
df_india = df_india.dropna()

# Extract years and unemployment values without specifying column names
years_india = df_india.columns[1:].values.astype(int)
unemployment_india = df_india.iloc[:, 1:].apply(pd.to_numeric, errors='coerce').values.flatten()

# Select data for first-world countries
first_world_countries = ['United States', 'Canada', 'United Kingdom', 'Germany', 'France', 'Japan', 'Australia']
df_first_world = df[df['Country Name'].isin(first_world_countries)]
df_first_world = df_first_world.dropna()

# Extract years and unemployment values without specifying column names
years_first_world = df_first_world.columns[1:].values.astype(int)
unemployment_first_world = df_first_world.iloc[:, 1:].apply(pd.to_numeric, errors='coerce').values

# Filter data for training on years 1990 to 2022
start_year_training = 1990
end_year_training = 2022
mask_training = (years_india >= start_year_training) & (years_india <= end_year_training)

# Apply the mask for both India and first-world countries
years_india_training = years_india[mask_training]
unemployment_india_training = unemployment_india[mask_training]

# Calculate the average unemployment rate for each year across first-world countries
avg_unemployment_first_world = np.mean(unemployment_first_world, axis=0)

# Set a larger plot size
plt.figure(figsize=(12, 6))

# Plot actual values for India
plt.plot(years_india, unemployment_india, label='Actual Unemployment (India)', marker='o', linestyle='-')

# Plot actual values for the average of first-world countries
plt.plot(years_first_world, avg_unemployment_first_world, label='Actual Avg. Unemployment (First World)', color='r', linestyle='--')

# ARIMA order selection using auto_arima or other methods

# ARIMA for India with different order
model_india = ARIMA(unemployment_india_training, order=(0, 0, 0))
fit_india = model_india.fit()
future_years_india = np.arange(2023, 2051)
predicted_unemployment_india = fit_india.forecast(steps=len(future_years_india))

# ARIMA for the average of first-world countries with different order
model_first_world = ARIMA(avg_unemployment_first_world, order=(2, 1, 0))
fit_first_world = model_first_world.fit()
predicted_unemployment_first_world = fit_first_world.forecast(steps=len(future_years_india))

# Print predicted values only
print("Predicted Unemployment (India):", predicted_unemployment_india)
print("Predicted Unemployment (First World):", predicted_unemployment_first_world)

# Calculate and print accuracy measures
mae_india = mean_absolute_error(unemployment_india[-len(future_years_india):], predicted_unemployment_india)
mse_india = mean_squared_error(unemployment_india[-len(future_years_india):], predicted_unemployment_india)

mae_first_world = mean_absolute_error(avg_unemployment_first_world[-len(future_years_india):], predicted_unemployment_first_world)
mse_first_world = mean_squared_error(avg_unemployment_first_world[-len(future_years_india):], predicted_unemployment_first_world)

print("MAE (India):", mae_india)
print("MSE (India):", mse_india)

print("MAE (First World):", mae_first_world)
print("MSE (First World):", mse_first_world)

# Plot predicted values for India
plt.plot(future_years_india, predicted_unemployment_india, label='ARIMA Prediction (India)', linestyle='--', color='blue')

# Plot predicted values for the average of first-world countries
plt.plot(future_years_india, predicted_unemployment_first_world, label='ARIMA Prediction (First World)', linestyle='--', color='green')

# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('Unemployment %')
plt.title('Unemployment Rate Comparison: India vs. First World'.format(mae_india, mse_india, mae_first_world, mse_first_world))
plt.legend()

# Increment x-axis ticks by 5
plt.xticks(range(1990, 2051, 5), rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Assuming your DataFrame is named df

# Select data for India
df_india = df[df['Country Name'] == 'India']
df_india = df_india.dropna()

# Extract years and unemployment values without specifying column names
years_india = df_india.columns[1:].values.astype(int)
unemployment_india = df_india.iloc[:, 1:].replace('NA', np.nan).astype(float).values.flatten()

# Split data into training and testing sets
train_size = int(len(years_india) * 0.8)
train_years, test_years = years_india[:train_size], years_india[train_size:]
train_unemployment, test_unemployment = unemployment_india[:train_size], unemployment_india[train_size:]

# Handle NaN values for ARIMA
valid_data_arima = ~np.isnan(train_unemployment)
train_years_arima, train_unemployment_arima = train_years[valid_data_arima], train_unemployment[valid_data_arima]
model_arima = ARIMA(train_unemployment_arima, order=(5, 1, 1))
fit_arima = model_arima.fit()

# Handle NaN values for Linear Regression
valid_data_linear = ~np.isnan(train_unemployment)
train_years_linear, train_unemployment_linear = train_years[valid_data_linear], train_unemployment[valid_data_linear]
model_linear = LinearRegression().fit(train_years_linear.reshape(-1, 1), train_unemployment_linear)

# Predict using ARIMA for the entire time period (1960 to 2050)
future_years = np.arange(2023, 2051, 1)
predicted_unemployment_arima = fit_arima.forecast(steps=len(future_years))

# Predict using Linear Regression for the entire time period (1960 to 2050)
predicted_unemployment_linear = model_linear.predict(future_years.reshape(-1, 1))

# Plot actual values for India
plt.plot(years_india, unemployment_india, label='Actual Unemployment (India)', marker='o', linestyle='-')

# Plot predicted values for ARIMA
plt.plot(future_years, predicted_unemployment_arima, label='ARIMA Prediction (India)', linestyle='--', color='blue')

# Plot predicted values for Linear Regression
plt.plot(future_years, predicted_unemployment_linear, label='Linear Regression Prediction (India)', linestyle='--', color='green')

# Calculate metrics for ARIMA
mae_arima = mean_absolute_error(unemployment_india[-len(future_years):], predicted_unemployment_arima)
mse_arima = mean_squared_error(unemployment_india[-len(future_years):], predicted_unemployment_arima)
rmse_arima = np.sqrt(mse_arima)
mape_arima = np.mean(np.abs((unemployment_india[-len(future_years):] - predicted_unemployment_arima) / unemployment_india[-len(future_years):])) * 100

# Calculate metrics for Linear Regression
mae_linear = mean_absolute_error(unemployment_india[-len(future_years):], predicted_unemployment_linear)
mse_linear = mean_squared_error(unemployment_india[-len(future_years):], predicted_unemployment_linear)
rmse_linear = np.sqrt(mse_linear)
mape_linear = np.mean(np.abs((unemployment_india[-len(future_years):] - predicted_unemployment_linear) / unemployment_india[-len(future_years):])) * 100

# Print metrics
print("\nARIMA Metrics:")
print("MAE: {:.4f}".format(mae_arima))
print("MSE: {:.4f}".format(mse_arima))
print("RMSE: {:.4f}".format(rmse_arima))
print("MAPE: {:.4f}%".format(mape_arima))

print("\nLinear Regression Metrics:")
print("MAE: {:.4f}".format(mae_linear))
print("MSE: {:.4f}".format(mse_linear))
print("RMSE: {:.4f}".format(rmse_linear))
print("MAPE: {:.4f}%".format(mape_linear))

# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('Unemployment %')
plt.title('Unemployment Rate Prediction Comparison: ARIMA vs. Linear Regression (India)')
plt.legend()

# Increment x-axis ticks by 5
plt.xticks(range(1990, 2051, 5), rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

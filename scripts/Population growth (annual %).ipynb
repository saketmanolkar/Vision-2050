```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Assuming 'df' is your DataFrame with relevant data
# Replace 'NA' with NaN
df.replace('NA', np.nan, inplace=True)

# Filter data for India
india_data = df[df['Country Name'] == 'India'].iloc[:, 1:].astype(float)

# Fill missing values using the mean
india_data_filled = india_data.fillna(india_data.mean())

# Set a larger figure size
plt.figure(figsize=(12, 6))

# Plotting the literacy rate for India (Actual)
years_actual_india = india_data_filled.columns.astype(int)
values_actual_india = india_data_filled.values.flatten()
plt.plot(years_actual_india, values_actual_india, marker='o', label='India (Actual)')

# Adding labels and title
plt.title('Population growth in India Over the Years')
plt.xlabel('Year')
plt.ylabel('Population growth')
plt.grid(True)

# ARIMA
order_arima = (2, 0, 0)  # You may need to adjust the order based on the characteristics of your data
model_arima = ARIMA(values_actual_india, order=order_arima)
fitted_model_arima = model_arima.fit()  # Use the default 'lbfgs' method
arima_predictions = fitted_model_arima.predict(start=len(values_actual_india), end=len(values_actual_india) + 27)  # Predict from 2023 to 2050

# Print predicted values for ARIMA
print("Predicted Values for the Years 2023 to 2050 (ARIMA):")
for year, prediction in zip(np.arange(2023, 2051), arima_predictions):
    print(f"Year {year}: {prediction}")

plt.plot(np.arange(2023, 2051), arima_predictions, label='ARIMA (Predicted)', linestyle='--')

# Drop NaN values for calculation
values_actual_india_no_nan = values_actual_india[~np.isnan(values_actual_india)]
fittedvalues_arima_no_nan = fitted_model_arima.fittedvalues[~np.isnan(fitted_model_arima.fittedvalues)]

# Align lengths of the two arrays
min_length = min(len(values_actual_india_no_nan), len(fittedvalues_arima_no_nan))
values_actual_india_no_nan = values_actual_india_no_nan[:min_length]
fittedvalues_arima_no_nan = fittedvalues_arima_no_nan[:min_length]

# Calculate and print R-squared for ARIMA
r2_arima = r2_score(values_actual_india_no_nan, fittedvalues_arima_no_nan)
print(f"R-squared for ARIMA: {r2_arima}")

# Calculate and print MSE for ARIMA
mse_arima = mean_squared_error(values_actual_india_no_nan, fittedvalues_arima_no_nan)
print(f"Mean Squared Error (MSE) for ARIMA: {mse_arima}")

# Calculate and print MAE for ARIMA
mae_arima = mean_absolute_error(values_actual_india_no_nan, fittedvalues_arima_no_nan)
print(f"Mean Absolute Error (MAE) for ARIMA: {mae_arima}")

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from statsmodels.tsa.arima.model import ARIMA

# Assuming 'df' is your DataFrame with relevant data
# Replace 'NA' with NaN
df.replace('NA', np.nan, inplace=True)

# Filter data for India
india_data = df[df['Country Name'] == 'India'].iloc[:, 1:].astype(float)

# Fill missing values using the mean
india_data_filled = india_data.fillna(india_data.mean())

# Set a larger figure size
plt.figure(figsize=(12, 6))

# Plotting the literacy rate for India (Actual)
years_actual_india = india_data_filled.columns.astype(int)
values_actual_india = india_data_filled.values.flatten()
plt.plot(years_actual_india, values_actual_india, marker='o', label='India (Actual)')

# Adding labels and title
plt.title('Population growth (annual %) - India - Linear Regression vs ARIMA')
plt.xlabel('Year')
plt.ylabel('Population growth')
plt.grid(True)

# Linear Regression
years_linear_regression = np.array(years_actual_india).reshape(-1, 1)
values_actual_india_no_nan = values_actual_india[~np.isnan(values_actual_india)]  # Remove NaN values for fitting
years_linear_regression_no_nan = years_linear_regression[~np.isnan(values_actual_india)]
model_linear = LinearRegression()
model_linear.fit(years_linear_regression_no_nan, values_actual_india_no_nan)

# Predicting for the actual years
linear_regression_predictions_actual = model_linear.predict(years_linear_regression_no_nan.reshape(-1, 1))

# Plotting Linear Regression predictions for the actual years
plt.plot(years_linear_regression_no_nan, linear_regression_predictions_actual, label='Linear Regression (Actual)', linestyle='--')

# Predicting for the future years (2023 to 2050)
years_future = np.arange(2023, 2051).reshape(-1, 1)
linear_regression_predictions_future = model_linear.predict(years_future)
plt.plot(np.arange(2023, 2051), linear_regression_predictions_future, label='Linear Regression (Predicted)', linestyle='--')

# Calculate and print R-squared for Linear Regression
r2_linear = model_linear.score(years_linear_regression_no_nan, values_actual_india_no_nan)
print(f"R-squared for Linear Regression: {r2_linear}")

# ARIMA
order_arima = (2, 0, 0)  # You may need to adjust the order based on the characteristics of your data
model_arima = ARIMA(values_actual_india, order=order_arima)
fitted_model_arima = model_arima.fit()  # Use the default 'lbfgs' method
arima_predictions = fitted_model_arima.predict(start=len(values_actual_india), end=len(values_actual_india) + 27)  # Predict from 2023 to 2050
plt.plot(np.arange(2023, 2051), arima_predictions, label='ARIMA (Predicted)', linestyle='--')

# Calculate and print R-squared for ARIMA
residuals_arima = values_actual_india - fitted_model_arima.fittedvalues
ssr_arima = np.sum(residuals_arima**2)
sst_arima = np.sum((values_actual_india - np.mean(values_actual_india))**2)
r2_arima = 1 - (ssr_arima / sst_arima)
print(f"R-squared for ARIMA: {r2_arima}")

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

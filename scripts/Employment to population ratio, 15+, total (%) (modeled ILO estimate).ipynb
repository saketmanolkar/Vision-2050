```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Assuming 'df' is your DataFrame with relevant data
# Replace 'NA' with NaN
df.replace('NA', np.nan, inplace=True)

# Filter data for India
india_data = df[df['Country Name'] == 'India'].iloc[:, 1:].astype(float)

# Drop rows with NaN values
india_data_cleaned = india_data.dropna(axis=1)

# Use data from 1990 to 2022 for training
train_data = india_data_cleaned.loc[:, '1990':'2022']

# Prepare data for linear regression
years = np.array(train_data.columns, dtype=int).reshape(-1, 1)
employment_ratio = train_data.values.flatten()

# Create and fit the linear regression model
model = LinearRegression()
model.fit(years, employment_ratio)

# Predict values for 2023 to 2050
future_years = np.arange(2023, 2051).reshape(-1, 1)
predicted_values = model.predict(future_years)

# Calculate R2
r2 = r2_score(employment_ratio, model.predict(years))

# Set a larger figure size
plt.figure(figsize=(12, 6))

# Plotting the original data
plt.plot(years.flatten(), employment_ratio, marker='o', label='India')

# Plotting the predicted values
plt.plot(future_years.flatten(), predicted_values, marker='o', label='Predicted')

# Adding labels and title
plt.title('Employment to Population Ratio - India: Actual vs Predicted (Linear Regression)')
plt.xlabel('Year')
plt.ylabel('Employment to Population Ratio')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()

# Print R2
print(f'R2: {r2}')
```

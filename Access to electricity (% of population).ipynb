```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with relevant data
# Replace 'NA' with NaN
df.replace('NA', np.nan, inplace=True)

# Filter data for India
india_data = df[df['Country Name'] == 'India'].iloc[:, 1:].astype(float)

# Drop rows with NaN values
india_data_cleaned = india_data.dropna(axis=1)

# Set a larger figure size
plt.figure(figsize=(12, 6))

# Plotting the literacy rate for India
years_actual_india = india_data_cleaned.columns.astype(int)
values_actual_india = india_data_cleaned.values.flatten()
plt.plot(years_actual_india, values_actual_india, marker='o', label='India (Actual)')

# Adding labels and title
plt.title('Access to electricity in India Over the Years')
plt.xlabel('Year')
plt.ylabel('Access to electricity (% of population)')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with relevant data
# Replace 'NA' with NaN
df.replace('NA', np.nan, inplace=True)

# Define regions and their countries
regions = {
    "Africa": africa_countries,
    "Northern Africa": northern_africa_countries,
    "Western Africa": west_africa_countries,
    "Eastern Africa": eastern_africa_countries,
    "Central Africa": central_africa_countries,
    "Southern Africa": southern_africa_countries,
    'Europe': europe_countries,
    "Northern Europe": northern_europe,
    "Western Europe": western_europe,
    "Eastern Europe": eastern_europe,
    "Southern Europe": southern_europe,
    'Asia': asia_countries,
    'Indian Subcontinent': indian_subcontinent_countries,
    'Middle East': middle_east_countries,
    'Central America': central_america_countries,
    'North America': north_america_countries,
    'South America': south_america_countries,
    'Oceania': oceania_countries
}

# Set a larger figure size
plt.figure(figsize=(12, 6))

for region, countries in regions.items():
    # Filter data for countries in the current region
    region_data = df[df['Country Name'].isin(countries)].iloc[:, 1:].astype(float)
    
    # Drop rows with NaN values
    region_data_cleaned = region_data.dropna(axis=1)

    # Plotting the access to electricity for the current region
    years_actual = region_data_cleaned.columns.astype(int)
    values_actual = region_data_cleaned.mean(axis=0)  # Use the mean value for each year
    plt.plot(years_actual, values_actual, marker='o', label=f'{region}')

# Adding labels and title
plt.title('Access to electricity Across Regions Over the Years')
plt.xlabel('Year')
plt.ylabel('Access to electricity (% of population)')
plt.grid(True)

# Place legend outside the graph
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with relevant data
# Replace 'NA' with NaN
df.replace('NA', np.nan, inplace=True)

# Selected countries
countries_of_interest = ['United States', 'Canada', 'United Kingdom', 'Germany', 'France', 'Japan', 'Australia', 'China']

# Filter data for selected countries
selected_data = df[df['Country Name'].isin(countries_of_interest)]
selected_data = selected_data.melt(id_vars='Country Name', var_name='Year', value_name='Access to electricity')
selected_data['Year'] = pd.to_numeric(selected_data['Year'], errors='coerce')

# Calculate the average for selected countries
average_selected_countries = selected_data.groupby('Year')['Access to electricity'].mean().reset_index()

# Filter data for India
india_data = df[df['Country Name'] == 'India'].iloc[:, 1:].astype(float)
india_data_cleaned = india_data.dropna(axis=1)

# Set a larger figure size
plt.figure(figsize=(12, 6))

# Plotting the average of selected countries
plt.plot(average_selected_countries['Year'], average_selected_countries['Access to electricity'], marker='o', label='Average of First World Countries')

# Plotting the access to electricity for India
years_actual_india = india_data_cleaned.columns.astype(int)
values_actual_india = india_data_cleaned.mean(axis=0)
plt.plot(years_actual_india, values_actual_india, marker='o', label='India (Actual)')

# Adding labels and title
plt.title('Access to electricity: First World Countries vs India Over the Years')
plt.xlabel('Year')
plt.ylabel('Access to electricity (% of population)')
plt.grid(True)

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
